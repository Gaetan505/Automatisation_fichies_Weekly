```python
import pandas as pd
import numpy as np

correctif = pd.read_excel("C:\\Users\\yahyaoui\\Documents\\Auto_Ecarts_prix\\SEM 11 Régularisation marges.xlsx",
                    sheet_name="correctif avoir à émettre")

# Supprimer les lignes où toutes les valeurs sont NaN ou 0
correctif = correctif.dropna(how="all").loc[~(correctif.fillna(0) == 0).all(axis=1)]

bon = pd.read_excel("C:\\Users\\yahyaoui\\Documents\\Auto_Ecarts_prix\\SEM 11 Régularisation marges.xlsx",
                    sheet_name="bon tarif facture à établir")

# Supprimer les lignes où toutes les valeurs sont NaN ou 0
bon = bon.dropna(how="all").loc[~(bon.fillna(0) == 0).all(axis=1)]

#reg = pd.concat([correctif, bon], ignore_index=True)
```

```python
# Définir les colonnes
colonnes = [
    "Depot","CODE CLIENT","Ref Client", "CODE IFLS","Quantité UVC","DPDF","Depot","CLIENT","IFLS","Validation",
    "Créer Script","Depot", "Client","IFLS","Quantité","Libellé","Message de validation","Message de validation"
]

# Créer un DataFrame vide avec ces colonnes
maquette = pd.DataFrame(columns=colonnes)
```

```python
maquette
```

```python
liste_entrepot=[]
liste_df=[]

for i in range(len(bon)):
    facture=bon['n° facture'].iloc[i]
    ifls=bon['ifls'].iloc[i]
    corr=correctif[(correctif['n° facture'] == facture) & (correctif['ifls'] == ifls)]
    if bon['total ht'].iloc[i] + corr['total ht'].values[0]!=0:
        entrepot=float(bon['entrepot'].iloc[i])
        if entrepot not in liste_entrepot:
            liste_entrepot.append(float(entrepot))
            liste_df.append(maquette.copy())
        index_entrepot = liste_entrepot.index(entrepot)       
        nouvelle_ligne = {"Depot": corr['entrepot'].values[0],
                          "CODE CLIENT": corr['N° client'].values[0],
                          "Ref Client": 'REGUPRIX',
                          "CODE IFLS": corr['ifls'].values[0],
                          "Quantité UVC": corr['quantité facturé'].values[0],
                          "CLIENT": corr['N° client'].values[0],
                          "IFLS": corr['ifls'].values[0],
                          "Client": corr['N° client'].values[0],
                          "Quantité": corr['quantité facturé'].values[0],
                          "Libellé": corr['Libellé'].values[0],
                          "DPDF": corr['montant ht'].values[0]} 
        liste_df[index_entrepot].loc[len(liste_df[index_entrepot])]=nouvelle_ligne
        nouvelle_ligne = {"Depot": bon['entrepot'].iloc[i],
                          "CODE CLIENT": bon['N° client'].iloc[i],
                          "Ref Client": 'REGUPRIX',
                          "CODE IFLS": bon['ifls'].iloc[i],
                          "Quantité UVC": bon['quantité facturé'].iloc[i],
                          "CLIENT": bon['N° client'].iloc[i],
                          "IFLS": bon['ifls'].iloc[i],
                          "Client": bon['N° client'].iloc[i],
                          "Quantité": bon['quantité facturé'].iloc[i],
                          "Libellé": bon['Libellé'].iloc[i],
                          "DPDF": bon['montant ht'].iloc[i] } 
        liste_df[index_entrepot].loc[len(liste_df[index_entrepot])]=nouvelle_ligne
        
for df in liste_df:
    df['Validation'] = np.where(df['CODE CLIENT'] == df['CODE CLIENT'].shift(-1), "NON", "VALID")
    df.sort_values(by=['CODE CLIENT', 'CODE IFLS'], ascending=[True, True], inplace=True)
    df['Message de validation']='Validation'
```
